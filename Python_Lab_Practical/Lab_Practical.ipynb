## (a) Given a list of words, return a dictionary where the key is the word and the value is the number of times it occurs in the list.

```python
def word_times(list_of_words):
    freq_dict = {}
    for word in list_of_words:
        word=word.lower()
        if word in freq_dict:
            freq_dict[word] += 1
        else:
            freq_dict[word] = 1
    return freq_dict

list_of_words = ["Ram", "harsh", "ram", "harsh", "krish", "ram"]
print("Given List Of Words:",list_of_words)
print("Resultant Dictionary Containing Key Is The Word And The Value Is The Number Of Times It Occurs In The List:",word_times(list_of_words))
```

## (b) Write a Python program to generate list with 'width'-bit gray code.Sample Output:['00', '01', '11', '10']  ['000', '001', '011', '010', '110', '111', '101', '100']

```python
def generate_gray(width):
    gray_codes = ["0", "1"]
    print("Given Gray Code:",gray_codes)
    
    for _ in range(width - 1):
        temp = []
        for code in gray_codes:
            temp.append("0" + code)
        for code in reversed(gray_codes):
            temp.append("1" + code)

        gray_codes = temp

    return gray_codes

w1 = 2
w2 = 3

print(f"Sample Output For Width {w1} Is:{generate_gray(w1)}")
print()
print(f"Sample Output For Width {w2} Is:{generate_gray(w2)}")
```

## (c) Write a Python program to find the first triangle number to have over n(given) divisors.A triangular number is calculated by the equation: n(n+1)/2

```python
def divisor_count(num):
    total = 0
    for i in range(1, int(num ** 0.5) + 1):
        if num % i == 0:
            if i == num // i:
                total += 1
            else:
                total += 2
    return total

def find_triangle_number(num):
    k = 1
    while True:
        triangle = k * (k + 1) // 2
        if divisor_count(triangle) > num:
            return triangle
        k += 1

num = int(input("Enter the number of divisors: "))
answer = find_triangle_number(num)
print(f"First triangle number with more than {num} divisors is: {answer}")
```
