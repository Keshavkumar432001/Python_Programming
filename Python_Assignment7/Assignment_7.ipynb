#### Q1 WAP that creates a list of numbers from 1 to 20 that are either divisible by 2 or divisible by 4.

```python
numbers = [x for x in range(1, 21) if x % 2 == 0 or x % 4 == 0]
print("Numbers divisible by 2 or 4:", numbers)
```

#### Q2 WAP to list a squares of number from 1 to 10 then use for  in construct to sum elements in list generated.

```python
squares = [x**2 for x in range(1, 11)]
total = 0
for num in squares:
    total += num
print("Squares of numbers from 1 to 10:", squares)
print("Sum of squares:", total)
```

#### Q3 WAP that defines a list of countries that are a member of BRICS, Check whether a country is a member of BRICKS or not.

```python
brics_countries = ["Brazil", "Russia", "India", "China", "South Africa"]
def is_brics_member(country):
    if country.title() in brics_countries:
        return f"{country.title()} is a member of BRICS."
    else:
        return f"{country.title()} is NOT a member of BRICS."
country_to_check = input("Enter the name of a country: ")
print(is_brics_member(country_to_check))
```

#### Q4 WAP that creates a list of 10 random integers, then create two lists odd list and Even list that has all odd and even values.

```python
import random
numbers = [random.randint(1, 100) for _ in range(10)]
even_list = [num for num in numbers if num % 2 == 0]
odd_list = [num for num in numbers if num % 2 != 0]
print("Original List:", numbers)
print("Even Numbers List:", even_list)
print("Odd Numbers List:", odd_list)
```

#### Q5 WAP that converts strings of all uppercase characters into strings of all lowercase characters using map() function.

```python
strings = ["HELLO", "WORLD", "PYTHON", "PROGRAMMING"]
lowercase_strings = list(map(str.lower, strings))
print("Original List:", strings)
print("Lowercase List:", lowercase_strings)
```

#### Q6 WAP to find largest value in a list using reduce() function.

```python
from functools import reduce
numbers = [23, 56, 12, 89, 34, 67, 90, 45]
largest = reduce(lambda a, b: a if a > b else b, numbers)
print("List:", numbers)
print("Largest Value:", largest)
```

#### Q7 WAP that scans an email address and forms a tuple of user name and domain.

```python
email = input("Enter your email address: ")
username, domain = email.split('@')
email_tuple = (username, domain)
print("Email Tuple:", email_tuple)
```

#### Q8 WAP that has a list of numbers both positive and negative, make a new tuple with only positive values from list.

```python
numbers = [10, -5, 7, -2, 0, 15, -8, 3]
positive_tuple = tuple(num for num in numbers if num > 0)
print("Original List:", numbers)
print("Tuple with Positive Values:", positive_tuple)
```

#### Q9 WAP to find Insertion, Union and Symmetrics difference between two sets: Coders and Analysts.
####     a. People working as coders as well as Analysts
####     b. People working as coders or Analysts
####     c. People working as Analysts but not coders
####     d. People working in only one of the groups

```python
coders = {"Anurag", "Rohan", "Rohit", "David"}
analysts = {"Rohit", "Shyam", "Aman", "Rohan"}

# a.Intersection
both_roles = coders.intersection(analysts)

# b.Union
coders_or_analysts = coders.union(analysts)

# c.Difference
analysts_not_coders = analysts.difference(coders)

# d.Symmetric Difference
only_one_group = coders.symmetric_difference(analysts)

print("a. Coders as well as Analysts:", both_roles)
print("b. Coders or Analysts:", coders_or_analysts)
print("c. Analysts but not Coders:", analysts_not_coders)
print("d. Only in one group:", only_one_group)
```
