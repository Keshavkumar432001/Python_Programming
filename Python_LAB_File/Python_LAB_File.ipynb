```python
## (a)
## Objective
## To calculate the area of a triangle using Heron’s formula based on the lengths of its three sides.
side_1 = float(input("Enter first side of triangle: "))
side_2 = float(input("Enter second side of triangle: "))
side_3 = float(input("Enter third side of triangle: "))
semi_perimeter = (side_1 + side_2 + side_3) / 2
# Apply Heron's formula
triangle_area = (semi_perimeter *
                 (semi_perimeter - side_1) *
                 (semi_perimeter - side_2) *
                 (semi_perimeter - side_3)) ** 0.5
print("Area of the triangle is:", triangle_area)

## Code Explanation
## The program takes the three sides of a triangle as input from the user and converts them to floating-point values.
## It calculates the semi-perimeter by adding the three sides and dividing the sum by 2.
## Heron’s formula is applied using the semi-perimeter and the side lengths to compute the area.
## The square root of the calculated expression gives the area of the triangle.
## Finally, the result is displayed using the print() function.
```

```python
## (a)
## Objective
## To remove all vowels from a given string entered by the user.
text = input("Enter a string: ")
vowels = "aeiouAEIOU"
result_text = ""
for ch in text:
    if ch not in vowels:
        result_text += ch
print("String without vowels:", result_text)
## Objective

To remove all vowels from a given string entered by the user.

## Code Explanation
## The program accepts a string input from the user.
## A string containing all uppercase and lowercase vowels is defined.
## An empty string is initialized to store the result.
## The program iterates through each character of the input string.
## Characters that are not vowels are added to the result string.
## Finally, the modified string without vowels is printed.
```

```python
## (b)
## Objective
## To check whether a given string is a palindrome by ignoring spaces, symbols, and letter case.
user_input = input("Enter a string: ")
clean_text = ""
for ch in user_input:
    if ch.isalnum():
        clean_text += ch.lower()
reversed_text = ""
for i in range(len(clean_text) - 1, -1, -1):
    reversed_text += clean_text[i]
if clean_text == reversed_text:
    print("The given string is a Palindrome")
else:
    print("The given string is NOT a Palindrome")
## Objective

To check whether a given string is a palindrome by ignoring spaces, symbols, and letter case.

## Code Explanation
## The program takes a string input from the user.
## It removes all non-alphanumeric characters and converts letters to lowercase to create a clean version of the string.
## A new string is formed by reversing the cleaned text using a loop.
## The cleaned string is compared with its reversed version.
## If both are the same, the string is identified as a palindrome; otherwise, it is not.
```

```python
## (c)
## Objective
## To reverse the order of words in a sentence entered by the user.
sentence = input("Enter a sentence: ")
words = sentence.split()
reversed_sentence = ""
for i in range(len(words) - 1, -1, -1):
    reversed_sentence += words[i] + " "
print("Reversed word order:", reversed_sentence.strip())
## Code Explanation
## The program accepts a sentence as input from the user.
## The split() method breaks the sentence into a list of individual words.
## A loop iterates through the list of words in reverse order using indices.
## Each word is appended to a new string with a space in between.
## The strip() method removes the extra space at the end before displaying the result.
```

## 	File Handling a.	Write a program that counts the number of tabs, spaces and newline characters in a file. b.	Write a program that copies first 10 bytes of a binary file into another. 

```python
## (a)
## Objective
## To count the number of tabs, spaces, and newline characters present in a text file.
file_name = input("Enter file name: ")
tab_count = 0
space_count = 0
newline_count = 0
with open(file_name, "r") as file:
    content = file.read()
for ch in content:
    if ch == "\t":
        tab_count += 1
    elif ch == " ":
        space_count += 1
    elif ch == "\n":
        newline_count += 1
print("Number of tabs      :", tab_count)
print("Number of spaces   :", space_count)
print("Number of new lines:", newline_count)
## Code Explanation
## The program asks the user to enter the name of a file.
## Three counters are initialized to keep track of tabs, spaces, and new lines.
## The file is opened in read mode and its entire content is read.
## Each character in the file content is examined one by one.
## Depending on the character type, the corresponding counter is increased.
## Finally, the total counts of tabs, spaces, and new lines are displayed.
```

```python
## (b)
## Objective
## To copy the first 10 bytes from one binary file to another binary file.
source_file = input("Enter source binary file name: ")
destination_file = input("Enter destination binary file name: ")
with open(source_file, "rb") as src:
    data = src.read(10)
with open(destination_file, "wb") as dest:
    dest.write(data)
print("First 10 bytes copied successfully.")
## Code Explanation
## The program asks the user to enter the source and destination binary file names.
## The source file is opened in binary read mode (rb).
## The read(10) method reads the first 10 bytes from the source file.
## The destination file is opened in binary write mode (wb).
## The read bytes are written into the destination file using write().
## A confirmation message is printed after the copy operation completes successfully.
```

## 	Programming Exercises with classes and objects, Inheritance a.	Write a program with class Employee that keeps a track of the number of     employees in an organization and also stores their name, designation and salary details.  b.
```python
## (a)
## Objective
## To demonstrate the use of a class, instance variables, class variables, and a class method in Python.
class Employee:
    employee_count = 0   
    def __init__(self, emp_name, emp_designation, emp_salary):
        self.name = emp_name
        self.designation = emp_designation
        self.salary = emp_salary
        Employee.employee_count += 1
    def show_details(self):
        print("Name        :", self.name)
        print("Designation :", self.designation)
        print("Salary      :", self.salary)
    @classmethod
    def total_employees(cls):
        return cls.employee_count
e1 = Employee("Rohit", "Manager", 55000)
e2 = Employee("Anita", "Developer", 42000)
print("Employee Details:\n")
e1.show_details()
print()
e2.show_details()
print("\nTotal Employees:", Employee.total_employees())
## Code Explanation
## The Employee class is defined with a class variable employee_count to track the total number of employees.
## The constructor (__init__) initializes employee details such as name, designation, and salary.
## Each time an object is created, the class variable employee_count is incremented.
## The show_details() method displays the details of an individual employee.
## The total_employees() class method returns the total number of employee objects created.
## Two employee objects are created, their details are displayed, and the total employee count is printed.
```

```python
## (b)
## Objective
## To model a simple student enrollment system using multiple classes and object relationships in Python.
class Department:
    def __init__(self, dept_name):
        self.name = dept_name
class Course:
    def __init__(self, course_name, course_year):
        self.name = course_name
        self.year = course_year
class Student:
    def __init__(self, student_name, roll_no, study_year):
        self.name = student_name
        self.rollno = roll_no
        self.year = study_year
        self.course = None
        self.department = None
    def enroll(self, course, department):
        self.course = course
        self.department = department
    def display_info(self):
        print("Student Name :", self.name)
        print("Roll Number  :", self.rollno)
        print("Year         :", self.year)
        print("Course       :", self.course.name)
        print("Department   :", self.department.name)
dept = Department("Computer Science")
course = Course("Python Programming", 2025)
stud = Student("Kiran", 101, "Second Year")
stud.enroll(course, dept)
stud.display_info()
## Code Explanation
## The Department class stores the name of a department.
## The Course class holds course details such as course name and academic year.
## The Student class stores student information and maintains references to a course and a department.
## The enroll() method assigns a course and department to a student.
## The display_info() method prints complete student details along with enrolled course and department.
## Objects for department, course, and student are created, the student is enrolled, and the details are displayed.
```

## 	Operator Overloading, Error Handling    a.	Write a program that overloads the + operator on a class Student that has attributes name and marks.   b.	Write a program that validates name and age as entered by the user to determine whether the person can cast vote or not.  c.	Write a program that asks the user to enter an integer.If the user enters something that is not an integer (like "abc" or 3.5), the program should catch the error, print "Invalid input, please enter an integer.", and ask again until a valid integer is entered.

```python
## (a)
## Objective
## To demonstrate operator overloading in Python by redefining the + operator for a user-defined class.
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def __add__(self, other):
        combined_name = self.name + " & " + other.name
        combined_marks = self.marks + other.marks
        return Student(combined_name, combined_marks)

    def display(self):
        print("Student Name:", self.name)
        print("Total Marks:", self.marks)
s1 = Student("Rahul", 78)
s2 = Student("Anita", 82)
result = s1 + s2
result.display()
## Objective

To demonstrate operator overloading in Python by redefining the + operator for a user-defined class.

## Code Explanation
## A Student class is defined with attributes for name and marks.
## The special method __add__() is implemented to overload the + operator.
## When two Student objects are added, their names are combined and their marks are summed.
## A new Student object is created and returned with the combined details.
## Two student objects are created and added using +.
## The display() method prints the details of the resulting student object.
```

```python
## (b)
## Objective
## To validate user input using exception handling and determine voting eligibility based on age.
try:
    name = input("Enter your name: ").strip()
    if not name.isalpha():
        raise ValueError("Name must contain only alphabets.")
    age = int(input("Enter your age: "))
    if age < 0:
        raise ValueError("Age cannot be negative.")
    if age >= 18:
        print(name, "is eligible to vote.")
    else:
        print(name, "is not eligible to vote.")
except ValueError as e:
    print("Error:", e)
Objective

To validate user input using exception handling and determine voting eligibility based on age.

## Code Explanation
## The program takes the user’s name and removes extra spaces using strip().
## It checks whether the name contains only alphabetic characters; otherwise, a ValueError is raised.
## The age is taken as an integer and validated to ensure it is not negative.
## Based on the age value, the program checks and prints voting eligibility.
## The try–except block catches any invalid input and displays an appropriate error message.
```
